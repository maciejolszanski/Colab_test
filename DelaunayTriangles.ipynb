```python
import numpy as np
import math
import scipy.spatial
import matplotlib.pyplot as plt

# ilosc punkt√≥w siatki
n = 600

szerokosc = 1
wysokosc = 1
```

```python
rzedy = math.floor(math.sqrt(n))
print(rzedy)
```

Siatka regularna

```python
xi = np.arange(0, szerokosc + szerokosc/(rzedy), szerokosc/(rzedy-1))
yi = np.arange(0, wysokosc + wysokosc/(rzedy), wysokosc/(rzedy-1))
print(xi)
print(yi)
```

```python
points = np.zeros((n,2))
for wiersz in range(0, rzedy):
  for kolumna in range(0,rzedy):
    points[wiersz*(rzedy) + kolumna, 0] = xi[kolumna]
    points[wiersz*(rzedy) + kolumna, 1] = yi[wiersz]
```

```python
tri = scipy.spatial.Delaunay(points)
plt.triplot(points[:,0], points[:,1], tri.simplices)
plt.show()
```
